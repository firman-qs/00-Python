```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
def fungsi(u,x):
    return (u[1],-2*u[1]-2*u[0]+np.cos(2*x))
y0 = [0,0]
xs = np.linspace(1,10,200)
us = odeint(fungsi,y0,xs)
ys = us[:,0]
```

```python
plt.plot(xs,ys,'-')
plt.plot(xs,ys,'r*')
plt.show()
```
