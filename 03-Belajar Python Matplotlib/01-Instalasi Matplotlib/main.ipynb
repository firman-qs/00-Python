````python
import numpy as np
import matplotlib.pyplot as plt
````

## Membuat lingkaran

````python
PI = np.pi
sudut = np.linspace(0, 2*PI, 50)
radius = 10

x = radius * np.cos(sudut)
y = radius * np.sin(sudut)

# inisiasi pyplot
plt.plot(x, y)
# tampilkan plot
plt.show()
````
