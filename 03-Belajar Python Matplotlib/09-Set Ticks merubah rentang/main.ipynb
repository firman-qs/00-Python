# Set Ticks

Import library

```python
import numpy as np
import matplotlib.pyplot as plt
```

Data

```python
sudut = np.linspace(0,360,20)
y = np.sin(np.deg2rad(sudut))
```

Plot

```python
plt.style.use('dark_background')

plot1 = plt.plot(sudut,y)

plt.title('Sinus Wave')
xlabel = plt.xlabel('Sudut')
ylabel = plt.ylabel('magnitude')
yticks = plt.yticks([1,0,0.5,-0.5,-1])
xticks = plt.xticks([0,45,90,135,180,225,270,315,360],
           [r'${angka}^o$'.format(angka=sudut) for sudut in [0,45,90,135,180,225,270,315,360]])

plt.setp(plot1, linewidth='1', color='y', marker='.', markersize='10')
plt.setp((xlabel, ylabel), fontsize='10')

plt.show()
```
