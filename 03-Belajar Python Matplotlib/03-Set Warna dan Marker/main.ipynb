# Set Warna dan Marker

```python
import numpy as np
import matplotlib.pyplot as plt

# grafik osilasi teredam pegas
def dampedOscillation(Amplitudo, m, b, w, t, phase = 0):
    horizontal = np.arange(0, t, 0.1)
    vertikal = Amplitudo*np.exp((-b*horizontal)/(2*m))*np.cos(w*horizontal+np.deg2rad(phase))
    return horizontal,vertikal

t, y = dampedOscillation(Amplitudo=20, b=1, m=2, w=2, t=15)
t1, y1 = dampedOscillation(Amplitudo=20, b=1, m=2, w=2, t=15, phase=90)

# Set Warna dan marker
plt.plot(t, y, 'r-o')
plt.plot(t1, y1, 'b--')
plt.show()
```
