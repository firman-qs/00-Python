# COPY LIST

```python
data_1 = ['Ucup', 'Otong', 'Dudung']
print(f'data_1 = {data_1}')
data_2 = data_1
print(f'data_2 = {data_2}')

# ubah 'data_2' (asumsi awal, 'data' tidak ikut berubah)
data_2[2] = 'Asep'
print(f'data_1 = {data_1}') # ternyata 'data' juga ikut berubah

# 'data_2' menggunakan memory (reference) yang sama dengan 'data'
print(f'address data_1: {hex(id(data_1))}')
print(f'address data_2: {hex(id(data_2))}')
```

```python
# Solusi nya adalah memisahkan reference, dengan menggunakan copy
data_1 = ['====', '====', '====']
data_2 = data_1
data_3 = data_1.copy()
print(f'data_1 = {data_1}')
print(f'data_2 = {data_2}')
print(f'data_3 = {data_3}')

data_2[1] = "####"
data_3[1] = "||||"

print(f'\ndata_1 = {data_1}, address: {hex(id(data_1))}')
print(f'data_2 = {data_2}, address: {hex(id(data_2))}')
print(f'data_3 = {data_3}, address: {hex(id(data_3))}')
```
